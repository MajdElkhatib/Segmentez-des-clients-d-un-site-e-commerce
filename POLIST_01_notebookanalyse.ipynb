```python
def calcul_distance(coord_customer, coord_seller):
    import math
    (lat_A, lng_A), (lat_B, lng_B) = coord_customer, coord_seller
# conversion en radians
    lat_X = math.radians(lat_A)
    lng_X = math.radians(lng_A)
    lat_Y = math.radians(lat_B)
    lng_Y = math.radians(lng_B)
    # Distance angulaire
    try:
        S_ab = math.acos(
            math.sin(lat_X) * math.sin(lat_Y) +
            math.cos(lat_X) * math.cos(lat_Y) * math.cos(lng_Y-lng_X))
        # Rayon de la Terre
        R_earth_Km = 6371
        # distance
        distance = S_ab * R_earth_Km
        return distance
    except Exception:
        pass
```

```python
def determination_coord(row):
    coord_customer = (
        row['lat_customer'],
        row['lng_customer'],
    )
    coord_seller = (
        row['lat_seller'],
        row['lng_seller'],
    )
    return calcul_distance(coord_customer, coord_seller)
```

```python
# Calculer la distance pour chaque client
data['distance_to_seller_km'] = data.apply(determination_coord, axis=1)
```

```python
# la fonction d'agrégation pour les colonnes booléennes
def agg_bool2(x): return (x == True).sum()
```

```python
# Calculer et ajouter des variables pertinentes
def customer_aggregation(customer):
    features_to_add = []

    # Ajouter nombre total d'articles
    features_to_add.append(data_customer
                           .size()
                           .rename('nb_purchased_item'))
    # Ajouter le total des produits distingués et commandés
    features_to_add.append(data_customer['product_id']
                           .nunique()
                           .rename('nb_distinct_product'))
    # Ajouter nombre de commandes
    features_to_add.append(data_customer['order_id']
                           .nunique()
                           .rename('nb_order'))
    # Ajouter prix total
    features_to_add.append(data_customer['price']
                           .sum()
                           .rename('sum_price'))
    # Ajouter somme de frets
    features_to_add.append(data_customer['freight_value']
                           .sum()
                           .rename('sum_freight_value'))
    # Ajouter somme de payments
    features_to_add.append(data_customer['payment_value']
                           .sum()
                           .rename('sum_payment'))
    # Ajouter moyen de paiement échelonnés
    features_to_add.append(data_customer['payment_installments']
                           .mean()
                           .rename('mean_payment_installment'))
    # Ajouter moyenne de scores
    features_to_add.append(data_customer['review_score']
                           .mean()
                           .rename('mean_review_score'))
    # Ajouter date de première commande
    features_to_add.append(data_customer['order_purchase_timestamp']
                           .min()
                           .rename('first_order_purchase_timestamp'))
    # Ajouter date de dernière commande
    features_to_add.append(data_customer['order_purchase_timestamp']
                           .max()
                           .rename('last_order_purchase_timestamp'))
    # Ajouter 'mean_distance_to_seller_km'
    features_to_add.append(data_customer['distance_to_seller_km']
                           .mean()
                           .rename('mean_distance_to_seller_km'))
    # Ajouter le delai de livraison
    features_to_add.append(data_customer['delivery_period']
                           .mean()
                           .rename('mean_delivery_period'))
    # Ajouter 'mean_product_volume_cm3'
    features_to_add.append(data_customer['product_volume_cm3']
                           .mean()
                           .rename('mean_product_volume_cm3'))
    # Ajouter 'mean_product_weight_g'
    features_to_add.append(data_customer['product_weight_g']
                           .mean()
                           .rename('mean_product_weight_g'))
    # Ajouter 'cash_payment_percentage'
    features_to_add.append((data_customer['cash_payment']
                           .agg(agg_bool2)
                           / data_customer.size())
                           .rename('cash_payment_percent'))
    # Ajouter'credit_card_percentage'
    features_to_add.append((data_customer['paid_credit_card']
                           .agg(agg_bool2)
                           / data_customer.size())
                           .rename('credit_card_percent'))
    # Ajouter'single_payment_sequential_percentage'
    features_to_add.append((data_customer['payment_sequential']
                           .agg(agg_bool2)
                           / data_customer.size())
                           .rename('single_paym_seq_percent'))
    # Merger
    for feature_series in features_to_add:
        customer = pd.merge(
            customer,
            feature_series,
            on='customer_unique_id')
    return customer
```

```python
# Fonction pour Calculer et afficher le taux de remplissage de colonnes

def columns_filling_rate(dataframe, columns='all', missing_only=False):
    
    import pandas as pd
    import numpy as np
    
    if columns == 'all':
        columns = dataframe.columns
        
    
    # initialisation des résultats
    results = pd.DataFrame(columns=['nb_values', 'missing_values', 'filling_rate'])
        
    for column in columns:

        # Nombre de valeurs sur chaque colonne
        values_count = dataframe[column].count()
        
        # compter missing values
        nb_rows = dataframe.shape[0]
        missing_values = nb_rows - values_count

        # Compter le taux de remplissage
        filling_rate = values_count / nb_rows
        if missing_only and missing_values == 0:
            filling_rate = np.nan
        
    
        results.loc[column] = [values_count, missing_values, filling_rate]

    # trier
    results = results.dropna(subset=['filling_rate'])
    results = results.sort_values('filling_rate')
    
    if results.empty == False:
        return results
    else:
        print("No missing value.")
```

```python
# A nouveau ajouter de variables pertinentes à
# notre jeu de données customer
def feature_engineering(customer):
    # supprimer 'customer_id' et 'customer_unique_id'
    customer = customer.drop(columns=[
        'customer_id'])
    

    # Ajouter la variable 'activity_period'
    customer['activity_period'] = pd.to_timedelta(
        customer['last_order_purchase_timestamp']
        - customer['first_order_purchase_timestamp'])

    # Ajouter la variable 'mean_period_between_orders'
    customer['mean_period_between_orders'] = pd.to_timedelta(
        customer['activity_period'] / customer['nb_order'])
    
    #customer['mean_period_between_orders'] = \
     #   customer['activity_period'] / customer['nb_order']  

    # Ajouter la variable 'mean_price_of_items'
    customer['mean_price_of_items'] =\
        customer['sum_price'] / customer['nb_purchased_item']

    # Ajouter la variable 'mean_price_of_orders'
    customer['mean_price_of_orders'] = \
        customer['sum_price'] / customer['nb_order']

    # Ajouter la variable 'mean_nb_items_per_order'
    customer['mean_nb_items_per_order'] = \
        customer['nb_purchased_item'] / customer['nb_order']

    # Ajouter la variable 'freight_value_percentage'
    customer['freight_value_percentage'] =\
        customer['sum_freight_value'] \
        / (customer['sum_price']
           + customer['sum_freight_value'])

    return customer


customer = feature_engineering(customer)
```

```python
# Création d'une variable "Région
def region_feature():

    regions_dict = {
        "North" : set(['AC', 'AM', 'RR', 'RO', 'PA', 'AP', 'TO']),
        "Northeast" : set(
            ['MA', 'PI', 'BA', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE']
        ),
        "Central" : set(['MT', 'GO', 'MS','DF']),
        "Southeast" : set(['MG', 'SP', 'ES', 'RJ']),
        "South" : set(['PR', 'SC', 'RS']),
    }

    def get_region(state):
        for region, set_states in regions_dict.items():
            if state in set_states:
                return region

    customer['customer_region'] =\
        customer['customer_state'].map(get_region)
    customer['customer_region'] = \
        customer['customer_region'].astype('category')
    customer['customer_region'].value_counts()


region_feature()
```
